# 基本工作流

1. 创建空场景
1. 添加物体
1. 添加光源
1. 添加相机
1. 将相机显示屏内容呈现

```python
# import packages

from glass_engine import *
from glass_engine.Geometries import *
from glass_engine.Lights import *
```

```python
# 创建空场景
scene = Scene()
# 添加几何体
sphere = Sphere()
sphere.position.z = 1
scene.add(sphere)
# 添加地板
floor = Floor()
scene.add(floor)
# 添加光源
light = DirLight()
light.pitch = -45
light.yaw = 45
scene.add(light)
# 添加相机
camera = Camera()
camera.position.z = 1.7
camera.position.y = -5
camera.pitch = -15
scene.add(camera)

camera.screen.show()
```
