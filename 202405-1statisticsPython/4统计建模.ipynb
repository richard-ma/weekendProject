# 统计建模

## 建模方法
* 响应变量：根据某个因素变化而变化（响应）的变量，例如啤酒销量
* 解释变量：对关注对象的变化进行解释的变量，例如天气气温影响啤酒销量

* 参数模型：尽量简化现象、用极少数参数表达的模型
* 非参数模型：不追求用尽量少的参数表达的模型（模型易于描述，但容易变得过分复杂失去价值）

* 线性模型：线性关系
    * 表达式：y = ax + b
    * b：截距
    * a：回归系数（系数、权重）
* 建模 = 模型选择 + 参数估计
* 参数估计可以由计算机完成
* 变量选择：选取解释变量的过程
    * 假设检验（讨论是否相关）
    * 信息量准则
        * 赤池信息量准则(Akaike Information Criterion, AIC) AIC越小模型越合适
* 空模型：没有解释变量的模型
* 模型评估：
    * 评估预测精度
    * 是否满足建模时所假设的条件

## 数据表示与模型名称
* 正态线性模型：假设响应变量服从正态分布的线性模型
* 回归分析（回归模型）：在正态线性模型中，解释变量为连续变量的模型
* 多元回归分析：含有多个解释变量的回归分析
* 方差分析（分类模型）：在正态线性模型中，解释变量为分类变量的模型
    * 当解释变量为一个种类时，叫一元方差分析（二个种类叫二元方差分析）
* 广义线性模型：响应变量未必服从正态分布的线性模型

## 参数估计：最大似然估计
* 似然：当参数为某值时抽到特定样本的概率（密度） Likelihood
* 似然函数：在给定参数时计算似然的函数
* 对数似然：似然取对数
    * 对数：单调递增、可将乘法转化为加法、绝对值不会过小
* 最大似然法：求使得似然或对数似然最大的参数，并把这个参数作为参数估计值的方法
* 最大似然估计量：通过最大似然法估计得到的参数
    * 最大对数似然：最大似然估计量对应的对数似然
* 多于参数：与问题没有直接关系的参数

## 参数估计：最小化损失
* 损失函数：用于使参数最小
* 残差：响应变量的实际值与通过模型预测的值之间的差 residual
    * 残差平方和：适合作为损失指标
* 最小二乘法：求使得残差平方和最小的参数，并把这个参数作为参数估计值的方法
* 最小二乘法和最大似然法参数估计结果相等
* 误差函数：在机器学习中，误差函数就是对数似然的相反数
* 分类模型（方差分析）不适合使用损失函数来进行参数估计

## 预测精度的评估与变量选择
* 拟合精度：模型与已知数据的契合度
* 预测精度：模型与位置数据的契合度
* 过拟合：拟合精度很高，预测精度却很低的现象
* 删除多余变量有可能提高预测精度
* 泛化误差：预测值和位置数据之间的误差
* 训练集
* 测试集
* 交叉验证：针对测试集评价模型预测精度的方法
    * 留出交叉验证：已知数据留出p个数据作为测试集，其余作为训练集
    * K折交叉验证：已知数据分为k组，取1组作为测试集，其余作为训练集，重复k次（所有数据组都做过测试集）
* 赤池信息量准则
    * AIC = -2 x （最大对数似然 - 参与估计的参数个数）
    * AIC越小，模型越合适
    * 增加的对数似然能否弥补更多的解释变量带来的缺点（可以使用AIC删除多余变量）
* 相对熵：统计模型的预测结果概率分布，和真正服从的分布比较，两者之间的差异
    * 最小化相对熵与平均对数似然
    * AIC与平均对数似然中的偏离
