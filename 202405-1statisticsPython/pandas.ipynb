# Pandas

```python
import numpy as np
import pandas as pd
```

## 数据帧

```python
array1 = np.arange(1, 6, 1)
array2 = np.arange(2, 11, 2)
array3 = ["A", "B", "C", "D", "E"]

df = pd.DataFrame({
    'col1': array1, 
    'col2': array2,
    'col3': array3,
})

df
```

```python
# 从csv文件中读取数据

csv_df = pd.read_csv("sample.csv")
csv_df
```

```python
# 连接数据帧

pd.concat([df, csv_df])
```

```python
pd.concat([df, csv_df], axis=1)
```

### 操作数据帧

```python
# 按列提取数据

csv_df['请假天数']
```

```python
# 提取数据帧的前几行

csv_df.head(3)
```

```python
# 用query函数查询数据行

csv_df.query('index==2')
```

```python
# 用query查询指定的列值

csv_df.query('请假天数==1')
```

```python
# 多条件query查询
# OR: |
# AND: &

csv_df.query('请假天数 == 1 & index < 13')
```
