## 各种概率分布
* 二值随机变量：只有两个值的随机变量
* 伯努利试验：得到两种结果中的一种的试验
* 成功概率：把得到的两种结果中的一种的概率
* 伯努利分布
* 二项分布：设成功概率为p，进行N次独立的伯努利试验，成功的次数m所服从的离散型概率分布

```python
import numpy as np
import pandas as pd
import scipy as sp
from scipy import stats

from matplotlib import pyplot as plt
import seaborn as sns
sns.set()

import statsmodels.formula.api as smf
import statsmodels.api as sm

%precision 3
%matplotlib inline
```

```python
stats.binom.pmf(k=1, n=2, p=0.5)
```

```python
np.random.seed(1)
stats.binom.rvs(n=10, p=0.2, size=5)
```

```python
binomial = stats.binom(n=10, p=0.2)

np.random.seed(1)
rvs_binomial = binomial.rvs(size=10000)

m = np.arange(0, 10, 1)
pmf_binomial = binomial.pmf(k=m)
sns.histplot(rvs_binomial, bins=m, kde=False, color='gray')
plt.plot(m, pmf_binomial, color='black')
```

## 泊松分布
* 用于计数型数据所服从的离散型概率分布

```python
stats.poisson.pmf(k=2, mu=5)
```

```python
np.random.seed(1)
stats.poisson.rvs(mu=2, size=5)
```

```python
poisson = stats.poisson(mu=2)

np.random.seed(1)
rvs_poisson = poisson.rvs(size=10000)

pmf_poisson = poisson.pmf(k=m)

sns.histplot(rvs_poisson, bins=m, kde=False, color='gray')

plt.plot(m, pmf_poisson, color='black')
```
